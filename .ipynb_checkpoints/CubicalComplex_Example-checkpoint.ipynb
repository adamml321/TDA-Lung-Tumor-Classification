# Generating a simple example of persistent homology with cubical complexes

```python
import numpy as np
import matplotlib.pylab as plt
import math
import gudhi as gd
```

```python
def BitmapToPhom(bitmap, imager):
    l,w = np.shape(bitmap)
    
    cubical = gd.CubicalComplex(dimensions = [l,w], top_dimensional_cells = bitmap.flatten())
    phom = cubical.compute_persistence()
    phom_0 = cubical.persistence_intervals_in_dimension(0)
    phom_1 = cubical.persistence_intervals_in_dimension(1)
    
    return(phom_0, phom_1)
```

```python
greyscale_img = np.array([[0.3, 0.3, 0.3, 0.3, 0.5],
                         [0.3, 0.9, 1, 0.9, 0.5],
                         [0.3, 0.9, 0.3, 0.7, 0.5],
                         [0.3, 0.9, 1, 0.9, 0.5],
                         [0.3, 0.1, 0.1, 0.1, 0.5]])

plt.imshow(greyscale_img, cmap = 'gray')
```

```python
#Show what it looks like for a feature to "die"
#Look at a range of filtration values

filter1 = greyscale_img > 0.65
filter2 = greyscale_img > 0.75
filter3 = greyscale_img > 0.95
filter4 = greyscale_img > 1


fig, (ax0, ax1, ax2, ax3, ax4) = plt.subplots(1, 5, figsize=(10, 5))
ax0.imshow(greyscale_img, cmap='gray')
ax0.set_title('Greyscale Image')
ax1.imshow(filter1, cmap='gray')
ax1.set_title('$\epsilon = 0.65$')
ax2.imshow(filter2, cmap='gray')
ax2.set_title('$\epsilon = 0.75$')
ax3.imshow(filter3, cmap='gray')
ax3.set_title('$\epsilon = 0.95$')
ax4.imshow(filter4, cmap='gray')
ax4.set_title('$\epsilon = 0.1$')
```
